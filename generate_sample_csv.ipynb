```python
import pandas as pd
```

```python
# Read input CSV
df = pd.read_excel('metadata/chipseq_metadata.xlsx', names=['srr_id', 'condition'])

# Create sample_id in format: CONDITION:SRR_ID
df['sample_id'] = df['condition'] + ':' + df['srr_id']

# Create fastq file path
df['fastq_path'] = 'fastq/' + df['srr_id'] + '.fastq'

# Select only the columns we need
output_df = df[['sample_id', 'fastq_path']]

# Write to CSV
output_df.to_csv('samples.csv', index=False)
```
